### Estadisticas de correlación

Coeficiente de correlation

Si la correlación es mayor a cero la correlación es positiva, si es menor a cero es negativa, si es cero no existe correlación.

P - Value

Si es mayor a 0.1 no hay certeza en el resultado, si es menor a 0.1 hay poca certeza, si es menor a 0.05 hay una certeza moderada y si al final es menor a 0.001 hay una fuerte certeza en el resultado.
