### Lidiando con valores faltantes

Estos valores pueden ser representados por un ?, N/A, O ó en blanco.

Se puede:
1- Eliminar los valores faltantes, eliminar la variable o la "data entry"
2- Reemplazar los valores faltantes con el promedio o un punto similar, o con la frecuencia u otra función.

dropna() es una función que elimina los valores nulos, sin embargo habría que revisar los valores que se eliminan, 
debido a que se pierde todo el registro.

Ejemplo :
df.dropna()

Una excelente forma de reemplazar es remplazar por ejemplo:

mean = df[x].mean()

df[x].replace(np.nan, mean)

Esto para remplazarlo por la media.

### Normalización de datos en Python

Es hacer uniforme los valores que poseen diferentes rangos

Una data normalizada porsee valores entre 0 y 1 despues de aplicarles diferentes tecnicas de normalización

Existen tres formas princiaples

a- Xnew = Xold/Xmax

b- Xnew = (Xold - Xmin) / (Xmax - Xmin)

c- Xnew = Xold - u / (desv estandar)
