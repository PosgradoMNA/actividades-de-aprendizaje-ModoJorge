### Importando y exportando Data en Python


```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

df = pd.read_csv(url , header = None)

df.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","Drive-wheels",
           "engine-location", "wheel-base", "lenght", "width", "height", "curb-weight", "engine-type", "num-of-cylinders",
          "engine-size", "fuel-system", "bore", "stroke", "compression-ratio","horsepower", "peak-rpm","city-mpg",
          "highway-mpg", "price"]
```

```python
df.columns = headers

df.head(5)
```

Hemos hecho un registro llamado headers, luego este registro o lista la hemos igualada a las columnas (encabezados), esto con el fin de colocar los encabezados reales de cada columna.

Recordemos que hay dos formas de importar información , desde una ruta o desde un formato como csv, json, excel, sql, etc.
